```python
data  = df_woinf.copy()
```

```python
data = data.rename(columns = {"PCOS (Y/N)":"Target"})
data
```

```python
data = data.drop(["Sl. No","Patient File No.","Unnamed: 44"],axis = 1)
```

```python
data.info()
```

```python
data["AMH(ng/mL)"].head()
```

```python
data["II    beta-HCG(mIU/mL)"].head()
```

```python
data["AMH(ng/mL)"] = pd.to_numeric(data["AMH(ng/mL)"], errors='coerce')
data["II    beta-HCG(mIU/mL)"] = pd.to_numeric(data["II    beta-HCG(mIU/mL)"], errors='coerce')
```

```python
data.columns[data.isnull().any()]
```

```python
data['Marraige Status (Yrs)'] = data['Marraige Status (Yrs)'].fillna(data['Marraige Status (Yrs)'].median())
data['II    beta-HCG(mIU/mL)'] = data['II    beta-HCG(mIU/mL)'].fillna(data['II    beta-HCG(mIU/mL)'].median())
data['AMH(ng/mL)'] = data['AMH(ng/mL)'].fillna(data['AMH(ng/mL)'].median())
data['Fast food (Y/N)'] = data['Fast food (Y/N)'].fillna(data['Fast food (Y/N)'].median())
```

```python
data.columns[data.isnull().any()]
```

```python
X = data.drop(labels = ["Target"],axis = 1)
y = data.Target
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,y, test_size=0.3,random_state=42)
```

```python
print("X_train",len(X_train))
print("X_test",len(X_test))
print("y_train",len(y_train))
print("y_test",len(y_test))
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
data
```

Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Acuratetea modelului Naive Bayes: {:.2f}".format(accuracy*100))
```

Logistic Regression

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Acuratetea modelului de regresie logistica: {:.2f}".format(accuracy*100))
```

Support Vector Machine

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
model = SVC()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Acuratetea modelului SVM: {:.2f}".format(accuracy*100))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("AcurateÈ›ea modelului Random Forest: {:.2f}".format(accuracy * 100))
```
